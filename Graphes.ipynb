```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array(['CLAD','CSEN','CSEN','LANG','MATH','MATH2','MECH','PHYS','VEDC'])
y = np.array([93.5,95,99.5,87,84,84,95,64,87])
plt.grid()
plt.plot(x, y,marker= '^')
plt.show()
plt.xlabel('subjects')
plt.ylabel('marks')
plt.bar(x,y)
plt.show()
plt.plot(x, y,'o:r')
plt.show()
plt.plot(x, y,'o')
plt.show()
a=['CLAD','CSEN','CSEN','LANG','MATH','MATH2','MECH','PHYS','VEDC']
plt.pie(y,labels=a)
plt.legend()
plt.show()
plt.hist(y)
plt.show()
plt.barh(x,y)
plt.show()
```
